```python
def Encode(zomato):
    for column in zomato.columns[~zomato.columns.isin(['rate','cost','votes'])]:
        zomato[column]=zomato[column].factorize()[0]
    return zomato
```

```python
zomato_en=Encode(zomato.copy())
```

```python
Y=pd.crosstab(zomato['rate'],zomato['book_table'])
```

```python
type(Y)
```

```python
Y.sum(1)
```

```python
Y.sum()
```

```python
Y.div(Y.sum(1).astype(float),axis=0).plot(kind='bar',color=['red','yellow'])
```

```python
plt.rcParams['figure.figsize']=(12,9)
Y.div(Y.sum(1).astype(float),axis=0).plot(kind='bar',color=['red','yellow'])
```

```python
plt.rcParams['figure.figsize']=(12,9)
Y.div(Y.sum(1).astype(float),axis=0).plot(kind='bar',color=['red','yellow'],stacked=True)
plt.title("Table Booking vs Normal Rate")
```

```python
plt.rcParams['figure.figsize']=(12,9)
Y.div(Y.sum(1).astype(float),axis=0).plot(kind='bar',color=['yellow','black'],stacked=True)
plt.title("Table Booking vs Normal Rate")
```

```python
sns.countplot(zomato['city'])
sns.countplot(zomato['city']).set_xticklabels(sns.countplot(zomato['city']).get_xticklabels(),
rotation=90,ha="right")
fig=plt.gcf()
fig.set_size_inches(14,15)
plt.title("Location wise count of Restaurants")
plt.show()
```

```python
loc_plt=pd.crosstab(zomato['rate'],zomato['city'])
loc_plt.plot(kind='bar',stacked=True)
fig=plt.gcf()
fig.set_size_inches(15,15)
plt.title('Location Wise Rating',fontsize=16,fontweight='bold')
plt.ylabel('Location',fontsize=10,fontweight='bold')
plt.xlabel('Rating',fontsize=10,fontweight='bold')
plt.xticks(fontsize=10,fontweight='bold')
plt.yticks(fontsize=10,fontweight='bold')
plt.legend(loc=2)
```

```python
loc_plt=pd.crosstab(zomato['rate'],zomato['rest_type'])
loc_plt.plot(kind='bar',stacked=True)
fig=plt.gcf()
fig.set_size_inches(15,15)
plt.title('Location Wise Rating',fontsize=16,fontweight='bold')
plt.ylabel('Location',fontsize=10,fontweight='bold')
plt.xlabel('Rating',fontsize=10,fontweight='bold')
plt.xticks(fontsize=10,fontweight='bold')
plt.yticks(fontsize=10,fontweight='bold')
plt.legend(loc=2).remove()
```
